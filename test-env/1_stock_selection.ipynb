```python
sp500 = yf.download(tickers, start=start_date, end=end_date)
```

```python
df_close = sp500['Close']
df_close.shape
```

```python
df_close.info(verbose=True)
```

```python
df_close.isnull().sum()
```

```python
daily_volatility    = df_close.pct_change().apply(lambda x: np.log(1+x)).std()     
weekly_volatility   = daily_volatility.pct_change().apply(lambda x: x*np.sqrt(5)) 
```

```python
WV = pd.DataFrame(
    weekly_volatility
).reset_index()
WV.columns = ["Tick", "Volatility"]
```

```python
sorted_weekly_volatility = WV.sort_values(by='Volatility', ascending=False)
```
